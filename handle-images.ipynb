```python
sjb = io.imread('../input/images/IMG_2656.jpg')
```

```python
plt.imshow(sjb)
```

```python
plt.imshow(255-sjb)
```

```python
def reverse_matrix(matrix):
    mirror_matrix = [row[::-1] for row in matrix]
    return mirror_matrix
def mirror(img):
    colors = []
    colors.append(img[:,:,0])
    colors.append(img[:,:,1])
    colors.append(img[:,:,2])
    
    for i in range(0,3):
        colors[i] = reverse_matrix(colors[i])
    
    mirror_img = np.dstack((colors[0],colors[1], colors[2])).astype(np.uint8)
    
    return mirror_img
        
    
```

```python
fliped_sjb = mirror(sjb)
```

```python
plt.imshow(fliped_sjb)
```

```python
colors = []
colors.append(sjb[:,:,0].T)
colors.append(sjb[:,:,1].T)
colors.append(sjb[:,:,2].T)
```

```python
# colors

t = np.dstack((colors[0],colors[1], colors[2])).astype(np.uint8)
t = mirror(t)
plt.imshow(t)
```
